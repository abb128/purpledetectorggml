```python
import torch
weights = torch.load('weights.pt', weights_only=True)

weights.keys()
```

```python
model = torch.nn.Sequential(
    torch.nn.Linear(64*3, 64),
    torch.nn.Sigmoid(),
    torch.nn.Linear(64, 1),
    torch.nn.Sigmoid()
)

model.load_state_dict(weights)
```

```python
import numpy as np
from PIL import Image
from matplotlib.pyplot import imshow

image = Image.open("test_dataset/purple/Dropped Image (6).png").convert('RGB').resize((64, 128))

result = model(torch.Tensor(np.array(image)).reshape(128, -1) / 255.0)
result = result.detach().numpy()

result = result.repeat(64).reshape(128, 64)
print(result.max())
print(result.shape)
imshow(
    np.array([(np.array(image)/255.0).sum(axis=-1)/3.0, result]).transpose(1, 0, 2).reshape(128, 128)
)
```

```python
import numpy as np
from PIL import Image
from matplotlib.pyplot import imshow

image = Image.open("test_dataset/nonpurple/Dropped Image (6).png").convert('RGB').resize((64, 128))

result = model(torch.Tensor(np.array(image)).reshape(128, -1) / 255.0)
result = result.detach().numpy()

print(result.sum() / 128)
result = result.repeat(64).reshape(128, 64)

print(result.shape)
imshow(
    np.array([(np.array(image)/255.0).sum(axis=-1)/3.0, result]).transpose(1, 0, 2).reshape(128, 128)
)
```
